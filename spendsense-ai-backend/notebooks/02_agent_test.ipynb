```python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain.agents import create_agent

# 1. Load your secret API keys
load_dotenv()

# 2. Give the AI a Tool (Web Search)
search_tool = TavilySearchResults(max_results=3)
tools = [search_tool]

# 3. Wake up the AI Brain
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

# 4. System Instructions
system_prompt = """You are AMRE (Agentic Merchant Resolution Engine). 
Your job is to investigate cryptic UPI IDs from bank SMS messages.
1. Use the search tool to look up the UPI ID.
2. Figure out the real-world company or brand name.
3. Determine the Business Category (e.g., Food, Shopping, Transport).
4. Provide a brief explanation of how you figured it out."""

# 5. Assemble the Agent (The modern LangChain v1.0 way!)
# 👇 THE FIX: Changing 'state_modifier' to 'system_prompt'
agent_executor = create_agent(llm, tools, system_prompt=system_prompt)

# 6. The Test Case
cryptic_upi_id = "qwikcilver@hdfcbank" 
print(f"🕵️‍♂️ Sherlock Agent investigating: {cryptic_upi_id}...\n")

# 7. Run the Agent!
response = agent_executor.invoke({"messages": [("user", f"Investigate this UPI ID: {cryptic_upi_id}")]})

# 8. Print the final conclusion cleanly!
raw_content = response["messages"][-1].content

# Check if Gemini returned a list, and pull out just the text
clean_text = raw_content[0]['text'] if isinstance(raw_content, list) else raw_content

print("\n--- 🎯 AGENT CONCLUSION ---")
print(clean_text)
```
